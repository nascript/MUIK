```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([100, 150, 200, 250, 300])
Y = np.array([200, 250, 300, 350, 400])

# Hitung rata-rata X dan Y
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Hitung koefisien kemiringan (m)
m_numerator = np.sum((X - mean_X) * (Y - mean_Y))
m_denominator = np.sum((X - mean_X)**2)
m = m_numerator / m_denominator

# Hitung intercept (b)
b = mean_Y - m * mean_X

# Persamaan regresi linear
equation = f"Y = {m:.4f}X + {b:.4f}"

# Prediksi jumlah poin loyalitas untuk X = 180
X_pred = 180
Y_pred = m * X_pred + b

# Plot data dan regresi linear
plt.scatter(X, Y, label='Data')
plt.plot(X, m * X + b, color='red', label='Regresi Linear Sederhana')
plt.scatter(X_pred, Y_pred, color='brown', label=f'Prediksi ({X_pred}, {Y_pred:.2f})')
plt.xlabel('Pengeluaran per Bulan ($) (X) independent')
plt.ylabel('Jumlah Poin Loyalitas (Y) dependen')
plt.title('Regresi Linear Sederhana')
plt.legend()
plt.show()

# Print hasil perhitungan
print(f"Rata-rata X: {mean_X}")
print(f"Rata-rata Y: {mean_Y}")
print(f"Koefisien kemiringan (m): {m_numerator} / {m_denominator} = {m:.4f}")
print(f"Intercept (b): {b:.4f}")
print(f"Persamaan regresi linear: {equation}")
print(f"Prediksi jumlah poin loyalitas untuk X = {X_pred}: Y ≈ {Y_pred:.2f}")
```
